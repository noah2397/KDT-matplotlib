```python
file="남북한발전전력량.xlsx"
```

- 실제 데이터 파일 확인 결과
* 첫 번째 행은 컬럼명 데이터
* 첫 번째와 두 번째 컬럼을 행 인덱스 사용으로 적합

    * 삭제하지 않고 대체로 처리 => 대체값 0

(3) 데이터 저장 : EXCEL => DataFrame

```python
korpowerDF=pd.read_excel(file, index_col=[0,1])
```

```python
korpowerDF.head()
```

(4) 데이터 확인

- (4-1) 데이터 타입과 실제 데이터 체크
    - '-' 데이터를 '0'으로 변경
    - 

```python
print(korpowerDF.index, korpowerDF.columns, sep='\n\n')
```

(5) 데이터 전처리  
- (5-1) 데이터 자료형 변환

```python
# '-' => '0'
korpowerDF=korpowerDF.replace('-', 0)
korpowerDF
```

```python
# object => int16
korpowerDF=korpowerDF.astype('int16')
```

```python
korpowerDF.info() # int16으로 잘 바뀌었다
```

(6) 데이터 분석 작업

    - 남북한의 수력발전량 비교하기

```python
waterDF=korpowerDF.xs('수력', level=1)
waterDF.shape
```
