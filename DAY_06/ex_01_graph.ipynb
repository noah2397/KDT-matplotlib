- [기본] 라인/선 그래프
    * 시간의 변화를 표현해주는 그래프
    * 연속적/순서 있는 데이터를 시각화할 경우 주로 사용됨

```python
# plot(data) => y축 데이터로 사용
# plot(data1, data2) => x축이 data1, y축이 dats2 데이터로 사용
# 반환값 : 라인/선 그래프 객체 리스트에 담겨서 반환됨
x=range(10)
y=[value ** 2 for value in x]
ret=plt.plot(x,y)
#파이썬에서 표준출력 기능 내장 함수 => print() 동일 역할
plt.show()
ret[0].get_xdata() 
```

- 라인/선 그래프에서 형식 즉 포맷지정 => str 타입의 포맷 설정
    * 색상-형태-선스타일 ==> 'ro--'(red, circle, dot-line)
    * 대표적으로 많이 사용되는 색상 : r - red, g - green, b - blue, c - cyan, m - magenta, y - yellow, k - black, w - white
    * 형태 : o - circle, ^ - triangle, s - square, v - pentagram, < - left triangle, > - right triangle, 8 - octagon, p - pentagram, * - star, h - hexagon, + - plus, x - x, D - diamond, d - thin_diamond, | - vline, _ - hline
    * 선스타일 : - - solid line, -- - dashed line, -. - dash-dot line, : - dot-dash line
    * 포맷 설정 방법 : plt.plot(x,y,'r--')

```python
ret=plt.plot(x,y,'p') # red circle이란 뜻이다
plt.show()
```

```python
ret=plt.plot(x,y,'ro--') # 줄로 연결이란 뜻이다
plt.show()
```

```python
ret=plt.plot(x,y,'r^--') # red triangle이란 뜻이다
plt.show()
```

```python
ret=plt.plot(x,y,'g^--') #초록색 삼각형으로 그린다
plt.show() # show는 print와 똑같이 작용한다
```

```python
x2=[i for i in range(1,100)]
y2=[value * 2 for value in range(1,100)]
ret=plt.plot(x,y,'g^--', x2, y2, 'ro--') #초록색 삼각형으로 그린다
```
